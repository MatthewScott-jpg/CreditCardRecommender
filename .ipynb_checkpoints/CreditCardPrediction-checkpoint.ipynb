```python
#Matthew Scott
#Credit card Data Analytics, classification, prediction of Customer trends
#https://www.kaggle.com/aryashah2k/credit-card-customer-data

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Credit_Card_Customer_Data.csv') 
```

```python
df.drop(['Sl_No','Customer Key'],axis=1).describe()
```

```python
df.corr(method='pearson')
```

```python
#Positive: credit limit+total cards, credit limit+online visits
#Negative: calls+cards, bank+online visits, bank+calls

df.plot(x="Avg_Credit_Limit", y="Total_visits_online", kind="scatter")
```

```python
#interactions are defined as any direct action connecting a customer with the bank
df['Interactions'] = df['Total_visits_bank'] + df['Total_visits_online'] + df['Total_calls_made']
```

```python
data = df.drop(['Sl_No','Customer Key'],axis=1).copy()
sns.pairplot(data)
#clusters found: 
#credit limit and online visits
#credit limit and cards held
#credit limit and total interactions
#online visits and cards held
#online visits and calls made
```

```python
i=1
plt.figure(figsize=(20,20))
for j in df.columns:
    plt.subplot(4,3,i)
    sns.histplot(data=df,x=df[j],kde=True)
    i+=1
```
